#**3) Exploratory Data Analysis (EDA) and Visualization**#
#*Campaign Performance:*

```python
# Plot a scatter plot to visualize the relationship between Acquisition_Cost and ROI
plt.figure(figsize=(9,6),facecolor='teal')
plt.scatter(data=df,x= 'Acquisition_Cost',y = 'ROI')
plt.title('Acquisition_Cost vs ROI')
plt.xlabel('Acquisition_Cost')
plt.ylabel('ROI')
plt.show()
```

```python
#Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type.

plt.figure(figsize=(12,7),facecolor='teal')
sns.barplot(data=df,x='Channel_Used',y='Conversion_Rate',hue='Campaign_Type',estimator ='mean', edgecolor ='black')
plt.title('Average Conversion_Rate by Channel_Used and Campaign_Type')
plt.xlabel('Channel_Used')
plt.ylabel('Average Conversion_Rate')
plt.show()
```

```python
#Create a bar chart to visualize the average Conversion_Rate for different Channel_Used, categorized by Campaign_Type
#using a box plot.

plt.figure(figsize=(10,7),facecolor='teal')
sns.boxplot(data=df,x='Campaign_Type',y='Engagement_Score' , color ='turquoise')
plt.title('Distribution of Engagement_Score by Campaign_Type')
plt.xlabel('Campaign_Type')
plt.ylabel('Engagement_Score')
plt.show()
```

```python
#Analyze the average ROI by Company using a bar chart to compare the profitability

plt.figure(figsize=(10,6),facecolor='teal')
ax =sns.barplot(data=df,x='Company',y='ROI',estimator='mean',color ='cyan',edgecolor='black')
ax.bar_label(ax.containers[0],label_type='center')
plt.title('Average R OI by Company')
plt.xlabel('Company')
plt.ylabel('Average ROI')
plt.show()
```

```python
#Examine the correlation between Engagement_Score and Conversion_Rate
#using a heatmap.

correlation_matrix = df[['Engagement_Score', 'Conversion_Rate']].corr()
plt.figure(figsize=(8,6))
plt.title('Correlation between Engagement_Score and Conversion_Rate')
plt.show(sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f"))
```

#*Customer Segmentation:*#

```python
#Create a count plot to visualize the distribution of Target_Audience.

plt.figure(figsize=(10,7) ,facecolor='teal')
ax = sns.countplot(data=df,x='Target_Audience',edgecolor='black')
ax.bar_label(ax.containers[0],label_type='edge')
plt.title('Distribution of Target_Audience')
plt.xlabel('Target_Audience')
plt.ylabel('Count')
plt.show()
```

```python
#Identify which Customer_Segment has the highest Conversion_Rate for each Language
#using a bar chart.

plt.figure(figsize=(10,6),facecolor='teal')
ax = sns.barplot(data=df,x='Customer_Segment',y='Conversion_Rate',hue='Language',edgecolor='black')
plt.title('Conversion_Rate by Customer_Segment and Language')
plt.xlabel('Customer_Segment')
plt.ylabel('Conversion_Rate')
plt.show()
```

```python
#Visualize the distribution of Acquisition_Cost across each Customer_Segment, categorized by Channel_Used.
#using a box plot.

plt.figure(figsize=(10,7),facecolor='teal')
sns.boxplot(data =df ,x ='Customer_Segment',y ='Acquisition_Cost',hue='Channel_Used')
plt.title('Distribution of Acquisition_Cost by Customer_Segment and Channel_Used')
plt.xlabel('Customer_Segment')
plt.ylabel('Acquisition_Cost')
plt.show()
```

```python
#the effectiveness of campaigns conducted in different languages.
#using a Bar Chart

plt.figure(figsize=(10,6),facecolor='teal')
ax=sns.barplot(data=df,x='Language',y='Conversion_Rate',estimator='mean',edgecolor='black')
ax.bar_label(ax.containers[0],label_type='center')
plt.title('Average Conversion_Rate by Language')
plt.xlabel('Language')
plt.ylabel('Average Conversion_Rate')
plt.show()
```

#*Customer Segmentation:*#

```python
#Compare the Engagement_Score for different Channels_Used, segmented by Campaign_Type.
#using a bar chart.

plt.figure(figsize=(10,6),facecolor='teal')
sns.barplot(data=df,x='Channel_Used',y='Engagement_Score',hue='Campaign_Type',edgecolor='black')
plt.title('Engagement_Score by Channel_Used and Campaign_Type')
plt.xlabel('Channel_Used')
plt.ylabel('Engagement_Score')
plt.show()
```

```python
#Show the distribution of total ROI across different Channels_Used
#using a pie chart.

total_ROI = df.groupby('Channel_Used')['ROI'].sum()
plt.figure(figsize=(7,7),facecolor='teal')
plt.pie(total_ROI,labels=total_ROI.index,autopct='%1.1f%%',startangle=90 ,colors = ['r','pink','aqua','gray','blue','gold'])
plt.title('Distribution of total ROI by Channel_Used')
plt.show()
```

```python
#Plot a scatter plot to show the relationship between Clicks and Impressions for each Channel_Used.

plt.figure(figsize=(10,6),facecolor='teal')
sns.scatterplot(data=df,x='Clicks',y='Impressions', hue='Channel_Used')
plt.title('Relationship between Clicks and Impressions by Channel_Used')
plt.xlabel('Clicks')
plt.ylabel('Impressions')
plt.show()
```

#*Time-Based Analysis:*#

```python
#Plot the distribution of Duration using a histogram.

plt.figure(figsize=(10,6),facecolor='teal')
ax=sns.histplot(data=df,x='Duration',bins=20,kde=True)
ax.bar_label(ax.containers[0],label_type='center')
plt.title('Distribution of Duration')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.show()
```

```python
##Analyze how the overall Conversion_Rate has changed over Date for each
#Company using a line chart.

plt.figure(figsize=(10,6),facecolor='teal')
sns.lineplot(data=df,x='Date',y='Conversion_Rate',hue='Company')
plt.title('Conversion_Rate Over Time by Company')
plt.xlabel('Date')
plt.ylabel('Conversion_Rate')
plt.show()
```

#*Geographic Analysis*#

```python

#Determine which location has the highest Acquisition_Cost using a bar chart.

plt.figure(figsize=(10,6),facecolor='teal')
ax=sns.barplot(data=df,x='Location',y='Acquisition_Cost',edgecolor='black')
plt.title('Acquisition_Cost by Location')
plt.xlabel('Location')
plt.ylabel('Acquisition_Cost')
plt.show()
```

```python
#Visualize the Conversion_Rate by different Location, categorized by Target_Audience.
#using a bar chart.

plt.figure(figsize=(10,6),facecolor='teal')
sns.barplot(data=df,x='Location',y='Conversion_Rate',hue='Target_Audience',edgecolor='black')
plt.title('Conversion_Rate by Location and Target_Audience')
plt.xlabel('Location')
plt.ylabel('Conversion_Rate')
plt.show()
```

```python
# The Proportion of ROI and location
# using a pie chart
location_roi = df.groupby('Location')['ROI'].sum()
plt.figure(figsize=(8, 8))
plt.pie(location_roi, labels=location_roi.index, startangle=90)
plt.title('Proportion of ROI by Location')
plt.show()
```
